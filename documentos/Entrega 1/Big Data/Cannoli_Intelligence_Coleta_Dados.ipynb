```python
import pandas as pd
```

```python
order = pd.read_csv("Order.csv", sep=";")
campaign = pd.read_csv("Campaign.csv", sep=";")
campaign_queue = pd.read_csv("CampaignQueue.csv", sep=";")
customer = pd.read_csv("Customer.csv", sep=";")

print('Bases carregadas!')
```

```python
# Tabela com dimensões de linha e coluna de cada base
datasets_info = pd.DataFrame({
    "Dataset": ["Order", "Campaign", "CampaignQueue", "Customer"],
    "Linhas": [order.shape[0], campaign.shape[0], campaign_queue.shape[0], customer.shape[0]],
    "Colunas": [order.shape[1], campaign.shape[1], campaign_queue.shape[1], customer.shape[1]]
})
display(datasets_info)
```

```python
# Exibe as primeiras 5 linhas de cada tabela
print("Order:")
display(order.head(5))

print()

print("\nCampaign:")
display(campaign.head(5))

print()

print("\nCampaignQueue:")
display(campaign_queue.head(5))

print()

print("\nCustomer:")
display(customer.head(5))
```

```python
# Dicionário com coluna -> tipo -> nulos -> únicos
def data_dictionary(df, name):
    dic = pd.DataFrame({
        "coluna": df.columns,
        "dtype": df.dtypes.astype(str),
        "nulos": df.isna().sum(),
        "unicos": df.nunique()
    })
    dic["dataset"] = name
    return dic
```

```python
# Dicionário da base Order
print("Order")
dict_order = data_dictionary(order, "Order")
display(dict_order)
print("\n\n")

# Dicionário da base Campaign
print("Campaign")
dict_campaign = data_dictionary(campaign, "Campaign")
display(dict_campaign)
print("\n\n")

# Dicionário da base Campaign Queue
print("CampaignQueue")
dict_campaign_queue = data_dictionary(campaign_queue, "CampaignQueue")
display(dict_campaign_queue)
print("\n\n")

# Dicionário da base Campaign Customer
print("Customer")
dict_customer = data_dictionary(customer, "Customer")
display(dict_customer)
```
